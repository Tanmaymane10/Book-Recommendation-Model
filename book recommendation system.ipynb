```python
# Initialize variables to store results
total_precision = 0
total_recall = 0
total_map = 0
total_mrr = 0
num_users = 0

# Iterate through each user in the testing set
for user_id in book_pivot.columns:
    # Get the index of the current user in the book_pivot DataFrame
    user_index = np.where(book_pivot.columns == user_id)[0][0]
    
    # Get the ratings given by the current user
    user_ratings = book_pivot.iloc[:, user_index]
    
    # Get the top-rated book for the current user
    top_book = user_ratings.idxmax()
    
    # Find k nearest neighbors for the top-rated book (here, k=6)
    distances, suggestions = model.kneighbors(book_pivot.loc[[top_book]], n_neighbors=6)
    
    # Extract the suggested books
    suggested_books = book_pivot.index[suggestions[0]][1:]  # Exclude the first suggestion (which is the top-rated book itself)
    
    # Evaluate precision and recall
    relevant_books = [book_id for book_id in user_ratings.index if book_id in suggested_books]
    precision = len(relevant_books) / 5  # Precision@5, as we recommended top 5 books
    recall = len(relevant_books) / len(user_ratings)  # Recall@5
    
    # Update total precision and recall
    total_precision += precision
    total_recall += recall
    
    # Calculate Mean Average Precision (MAP)
    map_score = sum([precision for book_id in relevant_books]) / len(relevant_books) if relevant_books else 0
    total_map += map_score
    
    # Calculate Mean Reciprocal Rank (MRR)
    mrr_score = 1 / (suggestions[0].tolist().index(top_book) + 1) if top_book in suggested_books else 0
    total_mrr += mrr_score
    
    num_users += 1

# Calculate average precision, recall, MAP, and MRR
avg_precision = total_precision / num_users
avg_recall = total_recall / num_users
avg_map = total_map / num_users
avg_mrr = total_mrr / num_users

print("Average Precision@5:", avg_precision)
print("Average MAP:", avg_map)
```
